```python
%matplotlib inline
import numpy as np
import pyJHTDB
import pyJHTDB.dbinfo
import matplotlib.pyplot as plt
import pyov2sgd
import time

from mpl_toolkits.mplot3d import axes3d
from matplotlib.pyplot import cm
from sklearn.cross_validation import ShuffleSplit
```

```python
N = 16
t = np.linspace(0, 2 * np.pi, N)
x = np.zeros((t.shape[0], t.shape[0], t.shape[0], 3), np.float32)
x[:, :, :, 0] = t[:, np.newaxis, np.newaxis]
x[:, :, :, 1] = t[np.newaxis, :, np.newaxis]
x[:, :, :, 2] = t[np.newaxis, np.newaxis, :]
```

```python
lJHTDB = pyJHTDB.libJHTDB()
lJHTDB.initialize()

T = pyJHTDB.dbinfo.isotropic1024coarse['time'][-1]
time_ = 0
u = lJHTDB.getData(
               time_,
               x,
               sinterp = 0,
               getFunction='getVelocity')
lJHTDB.finalize()
```

```python
fig = plt.figure(figsize=(16, 16))
ax = fig.gca(projection='3d')
# ax = fig.gca()

q = ax.quiver(x[:, :, :, 0].reshape(-1), x[:, :, :, 1].reshape(-1), x[:, :, :, 2].reshape(-1),
              u[:, :, :, 0].reshape(-1), u[:, :, :, 1].reshape(-1), u[:, :, :, 2].reshape(-1), 
              cmap=cm.jet, length=0.25, lw=2)
q.set_array(u[:, :, :, 0].reshape(-1) ** 2 + u[:, :, :, 1].reshape(-1) ** 2 + u[:, :, :, 2].reshape(-1) ** 2)
# plt.colorbar(fig)
```

```python
Xt = x.reshape((-1, 3)).copy()
yt = u.reshape((-1, 3)).copy()

eta0 = 1.
lbda = 0.0
nu = 0.
block = 200
cond = 1e-10
cap = -1
T = 10
p = 3

rs = ShuffleSplit(Xt.shape[0], 5, random_state=0, test_size=0.5)

err_tr = np.empty((5))
err_te = np.empty((5))
for (idx, (train_index, test_index)) in enumerate(rs):
    batch = Xt[train_index, :].shape[0]
    L = pyov2sgd.RidgeLoss()
    B = pyov2sgd.DecomposableGaussian(np.eye(p), 1.)
    # B = pyov2sgd.DecomposableSkewedChi2(np.eye(p), 20)
    C = pyov2sgd.InverseScaling(eta0, 0, 0)

    estimator = pyov2sgd.DSOVK(L, B, C, p, lbda, nu, T, batch, block, cap, cond)
    start = time.time()
    estimator.fit_dense(Xt[train_index, :], yt[train_index, :])
    stop = time.time()
    print(stop - start)
    pred = estimator.predict_dense(Xt[train_index, :])
    err_tr[idx] = np.mean((pred - yt[train_index, :]) ** 2)
    print(err_tr[idx])
    
    pred = estimator.predict_dense(Xt[test_index, :])
    err_te[idx] = np.mean((pred - yt[test_index, :]) ** 2)
    print(err_te[idx])
    
print(err_tr.mean())
print(err_te.mean())

# [ 0.12412183  0.25782862  0.13848606]
# [ 0.15823983  0.29582374  0.17142639]

# [ 0.11318623  0.24043508  0.12349865]
# [ 0.16126626  0.2933257   0.16890634]
```
