```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns

import numpy as np
import pyov2sgd
import time

from sklearn.preprocessing import StandardScaler
from sklearn.metrics import roc_auc_score

data = np.genfromtxt('/Users/Romain/Downloads/EEG Eye State.arff', skip_header=19, delimiter=',')
```

```python
s = StandardScaler()
data_t = s.fit_transform(data[:, :-1]).copy()
data_t = data_t[np.linalg.norm(data_t, axis=1) < 4, :]
data_t = s.fit_transform(data_t)

plt.figure(figsize=(16,5))
plt.subplot(2, 1, 1)
plt.plot(data_t[ :-1, :])
plt.vlines(4000, -4, 4, colors='r')
plt.ylim(-4, 4)

plt.subplot(2, 1, 2)
plt.pcolor(data[ :-1, -1].reshape(1, -1))
plt.vlines(4000, 0, 1, colors='r')
```

```python
X = data_t[ :3999, :].copy()
y = data[ :3999  , -1].copy().reshape(1, -1).T

p = 1

eta0 = 1.
lbda = 0.001
nu = 0.1
block = 1000
cond = 1e-10
cap = -1
T = 5
batch = X.shape[0]

L = pyov2sgd.SoftMaxLoss()
B = pyov2sgd.DecomposableGaussian(np.eye(p), .1)
# B = pyov2sgd.DecomposableSkewedChi2(np.eye(p), 20)
C = pyov2sgd.InverseScaling(eta0, 0, 0)

estimator = pyov2sgd.DSOVK(L, B, C, p, lbda, nu, T, 
                           batch, block, cap, cond)
start = time.time()
estimator.fit_dense(X, y)
stop = time.time()
print(roc_auc_score((data[ :3999  , -1] == 1).ravel(), estimator.predict_dense(data_t[ :3999, :]).ravel()))
print(roc_auc_score((data[ 4000:10000  , -1] == 1).ravel(), estimator.predict_dense(data_t[ 4000:10000, :]).ravel()))

plt.figure(figsize=(16,5))
plt.subplot(2, 1, 1)
plt.pcolor(estimator.predict_dense(data_t[ :3999, :]).T)

plt.subplot(2, 1, 2)
plt.pcolor(data[:3999, -1].reshape(1, -1))

plt.figure(figsize=(16,5))
plt.subplot(2, 1, 1)
plt.pcolor(estimator.predict_dense(data_t[ 4000:10000, :]).T)

plt.subplot(2, 1, 2)
plt.pcolor(data[4000:10000, -1].reshape(1, -1))
```

```python
X = data_t[ :3999, :].copy()
y = data_t[1:4000, :].copy()

p = X.shape[1]

eta0 = 1.
lbda = 0.00000001
nu = 0.
block = 100
cond = 1e-10
cap = -1
T = 1
batch = X.shape[0]

L = pyov2sgd.SoftMaxLoss()
B = pyov2sgd.DecomposableGaussian(np.eye(p), .1)
# B = pyov2sgd.DecomposableSkewedChi2(np.eye(p), 20)
C = pyov2sgd.InverseScaling(eta0, 0, 0)

estimator = pyov2sgd.DSOVK(L, B, C, p, lbda, nu, T, 
                           batch, block, cap, cond)
start = time.time()
estimator.fit_dense(X, y)
stop = time.time()
```

```python
eps = 1e-10
Jtr = np.zeros((p, p, 4000))
for t in range(4000):
    Jtr[:, :, t] = np.vstack(((estimator.predict_dense(data_t[t, :].reshape(1, -1) + eps * np.eye(p)[i, :]) - 
                               estimator.predict_dense(data_t[t, :].reshape(1, -1))) / eps for i in range(p)))
```

```python
plt.pcolor(np.mean(np.abs(Jtr), axis=2))
```

```python
Xn = Jtr.reshape((Jtr.shape[0] * Jtr.shape[1], Jtr.shape[2])).copy().T
yn = data[:4000, -1].copy().reshape(1, -1).T
```

```python
eta0 = 1.
lbda = 0.00001
nu = .1
block = 100
cond = 1e-10
cap = -1
T = 1
batch = Xn.shape[0]

L = pyov2sgd.SoftMaxLoss()
B = pyov2sgd.DecomposableGaussian(np.eye(1), 15.)
# B = pyov2sgd.DecomposableSkewedChi2(np.eye(p), 20)
C = pyov2sgd.InverseScaling(eta0, 0, 0)

estimator2 = pyov2sgd.DSOVK(L, B, C, 1, lbda, nu, T, 
                           batch, block, cap, cond)
start = time.time()
estimator2.fit_dense(Xn, yn)
stop = time.time()
print(roc_auc_score((yn == 1).ravel(), estimator2.predict_dense(Xn).ravel()))

plt.figure(figsize=(16,5))
plt.subplot(2, 1, 1)
plt.pcolor((estimator2.predict_dense(Jtr.reshape((Jtr.shape[0] * Jtr.shape[1], Jtr.shape[2]))) ).reshape(1, -1))

plt.subplot(2, 1, 2)
plt.pcolor(data[:4000, -1].reshape(1, -1))
```

```python
eps = 1e-10
J = np.zeros((p, p, 10000 - 4000))
for t in range(4000, 10000):
    J[:, :, t - 4000] = np.vstack(((estimator.predict_dense(data_t[t, :].reshape(1, -1) + eps * np.eye(p)[i, :]) - 
                                    estimator.predict_dense(data_t[t, :].reshape(1, -1))) / eps for i in range(p)))
```

```python
Xn = J.reshape((J.shape[0] * J.shape[1], J.shape[2])).T
yn = data[4000:10000, -1].copy().reshape(1, -1).T
print(roc_auc_score((yn == 1).ravel(), estimator2.predict_dense(Xn).ravel()))
```

```python
plt.figure(figsize=(16,5))
plt.subplot(2, 1, 1)
plt.pcolor(estimator2.predict_dense(Xn).reshape(1, -1))

plt.subplot(2, 1, 2)
plt.pcolor(data[4000:10000, -1].reshape(1, -1))
```
